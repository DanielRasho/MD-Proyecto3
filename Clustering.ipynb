# Determinando el número optimo de clusters

```python
# 🔹 Reducir la dimensionalidad con PCA antes de clustering

X = merged_df

# 🔹 Probar diferentes números de clusters
n_clusters = range(1, 11)  # De 1 a 10 clusters
bics = []
aics = []

for k in n_clusters:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(X)
    bics.append(gmm.bic(X))
    aics.append(gmm.aic(X))

# 🔹 Graficar la métrica BIC
plt.figure(figsize=(8, 5))
plt.plot(n_clusters, bics, marker='o', label="BIC", linestyle='dashed')
plt.plot(n_clusters, aics, marker='s', label="AIC", linestyle='dashed')
plt.xlabel("Número de Clusters")
plt.ylabel("Valor del BIC/AIC")
plt.title("Gráfica de Codo para Gaussian Mixture")
plt.legend()
plt.grid(True)
plt.show()
```

Usando el metodo del codo lo oportuno parece ser elegir 4 clusters

```python
k = 4
gmm = GaussianMixture(n_components=k, random_state=42)
clusters = gmm.fit_predict(X)

# 🔹 Calcular valores de silueta
silhouette_avg = silhouette_score(X, clusters)
sample_silhouette_values = silhouette_samples(X, clusters)

# 🔹 Crear gráfico de silueta
fig, ax = plt.subplots(figsize=(8, 6))
y_lower = 10
colors = cm.viridis(np.linspace(0, 1, k))

for i in range(k):
    # Seleccionar valores de silueta del cluster i
    ith_cluster_silhouette_values = sample_silhouette_values[clusters == i]
    ith_cluster_silhouette_values.sort()

    # Determinar tamaño del cluster
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # Rellenar el gráfico con los valores de silueta
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values,
                     facecolor=colors[i], edgecolor=colors[i], alpha=0.7)

    # Etiqueta para cada cluster en el eje Y
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Actualizar y_lower para el próximo cluster
    y_lower = y_upper + 10

# 🔹 Formato del gráfico
ax.set_title("Gráfico de Silueta para GMM con k=4")
ax.set_xlabel("Coeficiente de Silueta")
ax.set_ylabel("Clusters")
ax.axvline(x=silhouette_avg, color="red", linestyle="--", label="Promedio Silueta")
ax.set_yticks([])
ax.set_xlim([-0.1, 1])
ax.legend()
plt.show()

# 🔹 Mostrar el promedio del coeficiente de silueta
print(f"Coeficiente de silueta promedio: {silhouette_avg:.4f}")
```
